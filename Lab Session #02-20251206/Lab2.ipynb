```python
import pandas as pd
import matplotlib.pyplot as plt

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')
```

### Task 2: Read the CSV-files


```python
df = {}
for i in ["c7-1", "c7-2", "c8", "c9", "c11", "c13-1", "c13-2", "c14", "c15", "c16"]:
    df[i] = pd.read_csv(f"{i}.csv")
```

```python
df["c8"].head()
```

```python
# Get all column names (features) from c7-1
features = [col for col in df["c8"].columns]

# Calculate grid size
import math
n_features = len(features)
n_cols = 4  
n_rows = 7

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(30, n_rows * 3))
axes = axes.flatten()

# Plot each feature
for idx, feature in enumerate(features):
    df["c7-1"][feature].plot(ax=axes[idx], title=feature)
    axes[idx].set_xlabel('Index')
    axes[idx].set_ylabel(feature)



plt.tight_layout()
plt.show()
```
