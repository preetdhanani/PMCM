## Now for some plotting

```python
import matplotlib.pyplot as plt
```

```python
for f in X.columns:
    plt.plot(X[f],label=f)

plt.legend()
```

```python
cols = X.columns
cols = cols.drop("species")
for f in cols:
    plt.plot(X[f],label=f)

plt.legend()
```

```python
for sp in X.species.unique():
    xx = X[X.species == sp]
    for f in cols:
        plt.plot(xx.index - xx.index.min(),xx[f],label=f"{sp} - {f}")
#plt.legend()
```
