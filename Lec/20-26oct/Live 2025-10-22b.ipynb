# Regression on USA Housing data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv("../LectureNotes/Data/USA_Housing.csv")
```

```python
featureCols = ['Avg. Area Income', 
        'Avg. Area House Age', 
        'Avg. Area Number of Rooms',
        'Avg. Area Number of Bedrooms', 
        'Area Population', 
         #'Price', 
         #'Address'
        ]
targetCol = 'Price'

X = housing[featureCols]
y = housing[targetCol]
```

```python
X.describe()
```

```python
y.describe()
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(random_state=42)

reg.fit(X,y)

y_pred = pd.Series(reg.predict(X))
```

```python
plt.plot(y)
plt.plot(y_pred)
```

```python
plt.scatter(y_pred,y)
```
